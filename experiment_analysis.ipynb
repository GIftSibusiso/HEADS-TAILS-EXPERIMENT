# Tossing a coin 100 times, relative frequency explained

What happens when tossing a coin 100 times? In this experiment we'll be looking at indepth definition of relative frequency(also known as emperical/theoretical probabiity). When a fair coin is flipped it can land on either head or tail(2 mutually exclusive events) making the theoretical probability of the coin landing on head/tail 1/2 or 50%.

Relative frequency is equal to the number of times an event happens divided by the number of trials, e.g if I was to toss a coin 10 times and record the results as 7 heads and 3 tails the relative frequency of getting heads according to the results will be 7/10 which is 70%.
